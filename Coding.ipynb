```python
##########################################读取数据#############################################
import pandas as pd
data= pd.read_excel(r'F:/Ecoli/data/Inddata.xlsx',header=None)
s = data.iloc[:,0]
sequence = []
for i in  range(len(s)):
    a = ''.join(s[i])
    sequence.append(a)
#print(sequence)
```

```python
############################################BPF###############################################
bpf= {'A': [0,0,0,0],'G': [0,0,0,1],'C': [0,0,1,0],'T': [0,0,1,1]} 

import numpy as np
BPF=[]
for i in range(len(sequence)):
    seqcoding=[]
    base = []
    result=[]
    for j in sequence[i]:
        base.append(j)    
    for l in range(41):
        Bpf=bpf.get(base[l])
        seqcoding.append(Bpf)
    for j in range(len(seqcoding)):        
        for k in range(4):
            result.append(seqcoding[j][k]) 
    BPF.append(result)
print(BPF)
```

```python
############################################DPE###############################################
dpe= {'AA': [0,0,0,0],'AT': [0,0,0,1],'AC': [0,0,1,0],'AG': [0,0,1,1],
      'TA': [0,1,0,0],'TT': [0,1,0,1],'TC': [0,1,1,0],'TG': [0,1,1,1],
      'CA': [1,0,0,0],'CT': [1,0,0,1],'CC': [1,0,1,0],'CG': [1,0,1,1],
      'GA': [1,1,0,0],'GT': [1,1,0,1],'GC': [1,1,1,0],'GG': [1,1,1,1]} 

import numpy as np
DPE=[]
for i in range(len(sequence)):
    seqcoding=[]
    base = []
    dinucleo=[]
    result=[]
    for j in sequence[i]:
        base.append(j) 
    for k in range(len(base)-1):
        dinucleo.append(base[k]+base[k+1] ) 
    for l in range(40):
        Dpe=dpe.get(dinucleo[l])
        seqcoding.append(Dpe)
    for j in range(len(seqcoding)):        
        for k in range(4):
            result.append(seqcoding[j][k])
    DPE.append(result)
print(DPE)
```

```python
###########################################NPSE##############################################
eiip= {'A': 0.1260,'G': 0.0806,'C': 0.1340,'T': 0.1335} 

NPSE = []
for j in range(len(sequence)):
    seqcoding = []
    base = sequence[j]
    for i in range(len(base)):
        if i<=20:
            xi = 1*(21-i)/21*eiip.get(base[i])
        else:
            xi = 1*(i-19)/21*eiip.get(base[i])
        seqcoding.append(xi)
    result = []
    for m in seqcoding:
            result.append(m)
    NPSE.append(result)
print(NPSE)
```

```python
###########################################ENEI##############################################
K3=['AAA', 'AAT', 'AAC', 'AAG', 'ATA', 'ATT', 'ATC', 'ATG', 'ACA', 'ACT', 'ACC', 'ACG', 'AGA', 
    'AGT', 'AGC', 'AGG', 'TAA', 'TAT', 'TAC', 'TAG', 'TTA', 'TTT', 'TTC', 'TTG', 'TCA', 'TCT', 
    'TCC', 'TCG', 'TGA', 'TGT', 'TGC', 'TGG', 'CAA', 'CAT', 'CAC', 'CAG', 'CTA', 'CTT', 'CTC', 
    'CTG', 'CCA', 'CCT', 'CCC', 'CCG', 'CGA', 'CGT', 'CGC', 'CGG', 'GAA', 'GAT', 'GAC', 'GAG', 
    'GTA', 'GTT', 'GTC', 'GTG', 'GCA', 'GCT', 'GCC', 'GCG', 'GGA', 'GGT', 'GGC', 'GGG']
EIIP=[0.378, 0.3855, 0.386, 0.3326, 0.3855, 0.393, 0.3935, 0.3401, 0.386, 0.3935, 0.394, 0.3406, 0.3326, 0.3401, 
      0.3406, 0.2872, 0.3855, 0.393, 0.3935, 0.3401, 0.393, 0.4005, 0.401, 0.3476, 0.3935, 0.401, 0.4015, 0.3481,
      0.3401, 0.3476, 0.3481, 0.2947, 0.386, 0.3935, 0.394, 0.3406, 0.3935, 0.401, 0.4015, 0.3481, 0.394, 0.4015,
      0.402, 0.3486, 0.3406, 0.3481, 0.3486, 0.2952, 0.3326, 0.3401, 0.3406, 0.2872, 0.3401, 0.3476, 0.3481, 0.2947,
      0.3406, 0.3481, 0.3486, 0.2952, 0.2872, 0.2947, 0.2952, 0.2418]
dictionary=dict(zip(K3,EIIP))
print(dictionary)

ENEI=[]
for i in range(len(sequence)):
    seqcoding=[]
    base = []
    trinucleo=[]
    for j in sequence[i]:
        base.append(j)
    for k in range(len(base)-2):
        trinucleo.append(base[k]+base[k+1] +base[k+2])     
    for l in range(len(trinucleo)):
        result=[]
        Ni=1138 
        N=64
        s=((Ni)**0.5)/N
        for o in range(len(K3)):
            eiip=dictionary.get(K3[o])
            qqq=trinucleo[l].count(K3[o])
            p=((qqq+s)/(Ni+64*s))*eiip
            q=np.log(p)
            result.append(p*q)
        seqcoding.append(np.array(result).sum()*(-1))
    ENEI.append(seqcoding)
print(ENEI)
```

```python
###########################################NPSP##############################################
xi={'A': 1,'G': 1,'C': 0,'T': 0} 
yi={'A': 1,'G': 0,'C': 0,'T': 1}
zi={'A': 1,'G': 0,'C': 1,'T': 0}
NCPXS = []
for j in range(len(sequence)):
    seqcoding = []
    ccc = sequence[j]
    for i in range(len(ccc)):
        if i<=20:
            Xi=(21-i)/21*(xi.get(ccc[i]))
            Yi=(21-i)/21*(yi.get(ccc[i]))
            Zi=(21-i)/21*(zi.get(ccc[i]))
        else:
            Xi=(i-19)/21*(xi.get(ccc[i]))
            Yi=(i-19)/21*(yi.get(ccc[i]))
            Zi=(i-19)/21*(zi.get(ccc[i]))
        ncp=(Xi,Yi,Zi)    
        seqcoding.append(ncp)
    result = []
    for m in seqcoding:
        for n in m:
            result.append(n)
    NCPXS.append(result)
print(NCPXS)
```
